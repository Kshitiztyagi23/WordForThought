```python
import librosa
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from librosa.feature import melspectrogram
import itertools

# Step 3: Define the path for CREMA-D
crema_path = '/root/.cache/kagglehub/datasets/ejlok1/cremad/versions/1/AudioWAV'
print(f"CREMA-D Path: {crema_path}")

# Step 4: Check if the directory exists
if os.path.exists(crema_path):
    print(f"CREMA-D directory exists at: {crema_path}")
else:
    print(f"CREMA-D path does not exist. Check the download process.")

# Listing contents of CREMA-D
print("\nListing the contents of CREMA-D:")
print(os.listdir(crema_path) if os.path.exists(crema_path) else "CREMA-D directory not found.")

# Step 5: Feature Extraction Function
def extract_features(file_path, max_len=128):  # Added max_len for padding/truncation
    try:
        # Load the audio file
        data, sample_rate = librosa.load(file_path, duration=2.5, offset=0.6)

        # Extract Mel spectrogram
        mel_spect = librosa.feature.melspectrogram(y=data, sr=sample_rate)

        # Convert to log scale (log-mel spectrogram)
        log_mel_spect = librosa.power_to_db(mel_spect, ref=np.max)

        # Flatten the spectrogram and pad/truncate to the same length
        flattened = log_mel_spect.flatten()

        # If the feature is too short, pad it with zeros
        if len(flattened) < max_len:
            padded = np.pad(flattened, (0, max_len - len(flattened)), 'constant')
        # If the feature is too long, truncate it
        elif len(flattened) > max_len:
            padded = flattened[:max_len]

        return padded

    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

# Step 6: Process CREMA-D Dataset for Training and Testing
def process_crema_dataset(path, max_len=128):
    features = []
    labels = []

    for file_name in os.listdir(path):
        if file_name.endswith(".wav"):
            file_path = os.path.join(path, file_name)
            feature = extract_features(file_path, max_len)

            if feature is not None:
                features.append(feature)
                # Assuming label is in the first part of the filename (e.g., 'angry_01.wav')
                label = file_name.split('_')[2]  # Get the emotion label (e.g., 'ANG')
                labels.append(label)

    print(f"Processed {len(features)} samples from CREMA-D.")
    return np.array(features), np.array(labels)

# Step 7: Extract Features and Labels from CREMA-D
X, y = process_crema_dataset(crema_path, max_len=128)

# Step 8: Encode Labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Step 9: Split the Data into Training and Validation Sets
X_train, X_val, y_train_encoded, y_val_encoded = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Step 10: Standardize the Features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

# Step 11: Train Logistic Regression Model
lr_model = LogisticRegression(max_iter=1000, random_state=42)
lr_model.fit(X_train, y_train_encoded)

# Step 12: Make Predictions and Evaluate Model with Logistic Regression
lr_predictions = lr_model.predict(X_val)
lr_predictions_decoded = label_encoder.inverse_transform(lr_predictions)

# Calculate accuracy
lr_accuracy = accuracy_score(y_val_encoded, lr_predictions)
print(f"Logistic Regression Accuracy: {lr_accuracy:.4f}")

# Generate classification report for Logistic Regression
lr_report = classification_report(y_val_encoded, lr_predictions, output_dict=True, zero_division=1)
print(pd.DataFrame(lr_report).transpose())

# Step 13: Confusion Matrix Visualization for Logistic Regression
cm_lr = confusion_matrix(y_val_encoded, lr_predictions)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix for Logistic Regression')
plt.show()

# Step 14: Train XGBoost Model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(X_train, y_train_encoded)

# Step 15: Make Predictions and Evaluate Model with XGBoost
xgb_predictions = xgb_model.predict(X_val)
xgb_predictions_decoded = label_encoder.inverse_transform(xgb_predictions)

# Calculate accuracy for XGBoost
xgb_accuracy = accuracy_score(y_val_encoded, xgb_predictions)
print(f"XGBoost Accuracy: {xgb_accuracy:.4f}")

# Generate classification report for XGBoost
xgb_report = classification_report(y_val_encoded, xgb_predictions, output_dict=True, zero_division=1)
print(pd.DataFrame(xgb_report).transpose())

# Step 16: Confusion Matrix Visualization for XGBoost
cm_xgb = confusion_matrix(y_val_encoded, xgb_predictions)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix for XGBoost')
plt.show()

# Step 17: Feature Importance Visualization for XGBoost
plt.figure(figsize=(12, 10))
from xgboost import plot_importance
plot_importance(xgb_model, max_num_features=10, importance_type='weight')
plt.title('Feature Importance for XGBoost')
plt.show()

```

```python
import os
import librosa
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.utils.class_weight import compute_class_weight
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold

# Step 1: Define the CREMA-D dataset path
crema_path = '/root/.cache/kagglehub/datasets/ejlok1/cremad/versions/1/AudioWAV'

# Step 2: Function to extract features (MFCC, Chroma, Spectral Contrast, Mel-spectrogram)
def extract_features(audio_path):
    try:
        y, sr = librosa.load(audio_path, sr=None)
        # Mel-spectrogram
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=8000)
        mel_db = librosa.power_to_db(mel, ref=np.max)

        # MFCC
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

        # Chroma
        chroma = librosa.feature.chroma_stft(y=y, sr=sr)

        # Spectral Contrast
        spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr)

        # Combine features into one array (flattened)
        features = np.hstack([np.mean(mel_db, axis=1), np.mean(mfcc, axis=1), np.mean(chroma, axis=1), np.mean(spectral_contrast, axis=1)])
        return features
    except Exception as e:
        print(f"Error processing {audio_path}: {e}")
        return None

# Step 3: Function to process the CREMA-D dataset
def process_crema_dataset(path):
    features = []
    labels = []
    for file_name in os.listdir(path):
        if file_name.endswith('.wav'):
            file_path = os.path.join(path, file_name)
            feature = extract_features(file_path)
            if feature is not None:
                features.append(feature)
                # Extract label (emotion) from filename
                label = file_name.split('_')[2]  # Assuming the label is the 3rd part of the filename (e.g., 'HAP' or 'SAD')
                labels.append(label)
    return np.array(features), np.array(labels)

# Step 4: Extract Features and Labels from CREMA-D
X, y = process_crema_dataset(crema_path)

# Step 5: Encode Labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Step 6: Split the Data into Training and Validation Sets
X_train, X_val, y_train, y_val = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Step 7: Scale the Data (Logistic Regression requires scaled data)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Step 8: Handle Class Imbalance using Class Weights (optional)
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(enumerate(class_weights))

# Step 9: Train Models and Evaluate Accuracy

# Logistic Regression
log_reg = LogisticRegression(max_iter=2000, class_weight='balanced', random_state=42)
log_reg.fit(X_train_scaled, y_train)
y_pred_log_reg = log_reg.predict(X_val_scaled)
log_reg_accuracy = accuracy_score(y_val, y_pred_log_reg)
print("Logistic Regression Accuracy: {:.4f}".format(log_reg_accuracy))
print(classification_report(y_val, y_pred_log_reg))

# XGBoost Model
xgb_model = XGBClassifier(eval_metric='mlogloss', random_state=42)  # Removed invalid parameters
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_val)
xgb_accuracy = accuracy_score(y_val, y_pred_xgb)
print("XGBoost Accuracy: {:.4f}".format(xgb_accuracy))
print(classification_report(y_val, y_pred_xgb))

# Random Forest Classifier (Alternative Model)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_val)
rf_accuracy = accuracy_score(y_val, y_pred_rf)
print("Random Forest Accuracy: {:.4f}".format(rf_accuracy))
print(classification_report(y_val, y_pred_rf))

# Support Vector Classifier (Alternative Model)
svc_model = SVC(class_weight='balanced', random_state=42)
svc_model.fit(X_train, y_train)
y_pred_svc = svc_model.predict(X_val)
svc_accuracy = accuracy_score(y_val, y_pred_svc)
print("SVC Accuracy: {:.4f}".format(svc_accuracy))
print(classification_report(y_val, y_pred_svc))

# Step 10: Cross-Validation Scores for Logistic Regression
cv_scores_log_reg = cross_val_score(log_reg, X, y_encoded, cv=5)
print(f"Logistic Regression Cross-validation Accuracy: {cv_scores_log_reg.mean():.4f}")

# Step 11: Cross-Validation for XGBoost using Stratified KFold
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
xgb_cv_scores = []
for train_index, val_index in kf.split(X, y_encoded):
    X_train_cv, X_val_cv = X[train_index], X[val_index]
    y_train_cv, y_val_cv = y_encoded[train_index], y_encoded[val_index]
    xgb_model.fit(X_train_cv, y_train_cv)
    y_pred_cv = xgb_model.predict(X_val_cv)
    xgb_cv_scores.append(accuracy_score(y_val_cv, y_pred_cv))

print(f"XGBoost Cross-validation Accuracy: {np.mean(xgb_cv_scores):.4f}")
```
