Reshaping the Y arrays into 2D.

```python
y_train=y_train.reshape(1354,1)
y_tets=y_test.reshape(590,1)
```

Importing model and optimizers.

```python
from tensorflow.keras import Sequential

from tensorflow.keras.layers import Dense,Conv2D,MaxPool2D,Flatten,Dropout,BatchNormalization
from tensorflow.keras.optimizers import Adam
```

The architecture

```python
model=Sequential()

model.add(Conv2D(64,activation='relu',kernel_size=(3,3),input_shape=X_train[0].shape))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(15))
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

Training

```python
model.fit(X_train, y_train, epochs=10)
```

The accuracy

```python
test_loss, test_acc = model.evaluate(X_test,  y_test, verbose=2)
print('Test accuracy:', test_acc)
```

Changing into a probability model.

```python
probability_model = tf.keras.Sequential([model,tf.keras.layers.Softmax()])
```

```python
predictions = probability_model.predict(X_test)
```
