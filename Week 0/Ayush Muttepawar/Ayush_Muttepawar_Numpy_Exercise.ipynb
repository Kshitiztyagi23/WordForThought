## Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

```python
arr=np.arange(1,26).reshape(5,5)
print(arr)
```

```python
print(arr[2:,1:])
```

```python
print(arr[3,4])
```

```python
print(arr[:3,1:2])
```

```python
print(arr[4,:5])
```

```python
print(arr[3:5,:5])
```
